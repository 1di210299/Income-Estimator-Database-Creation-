```python
# importamos pandas y el documento del 2019
import pandas as pd
import numpy as np

caractericas_vivienda_hogar_2019 = pd.read_csv('../Data 2019/Características de la Vivienda y del Hogar/Enaho01-2019-100.csv', encoding='latin1')


caractericas_vivienda_hogar_2019

caractericas_vivienda_hogar_2019 = caractericas_vivienda_hogar_2019[['UBIGEO', 'P110A1', 'P1121', 'P1141', 'P1142', 'P1143', 'P1144']]

caractericas_vivienda_hogar_2019 = caractericas_vivienda_hogar_2019.rename(columns={'ubigeo':'departamento','P110A1': 'agua_potable', 
                              'P1121': 'electricidad',
                              'P1141': 'telefono',
                              'P1142': 'celular',
                              'P1144': 'internet',
                              'P1143':'tv'})

caractericas_vivienda_hogar_2019 = caractericas_vivienda_hogar_2019.rename(columns=str.lower)

#departamento
departamentos = {
    "01": "Amazonas",
    "02": "Áncash",
    "03": "Apurímac",
    "04": "Arequipa",
    "05": "Ayacucho",
    "06": "Cajamarca",
    "07": "Callao",
    "08": "Cusco",
    "09": "Huancavelica",
    "10": "Huanuco",
    "11": "Ica",
    "12": "Junin",
    "13": "La Liberta",
    "14": "Lambayeque",
    "15": "Lima",
    "16": "Loreto",
    "17": "Madre de Dios",
    "18": "Moquegua",
    "19": "Pasco",
    "20": "Piura",
    "21": "Puno",
    "22": "San Martin",
    "23": "Tacna",
    "24": "Tumbes",
    "25": "Ucayali"
}
caractericas_vivienda_hogar_2019['ubigeo'] = caractericas_vivienda_hogar_2019['ubigeo'].astype(str)

caractericas_vivienda_hogar_2019['COD_DEPARTAMENTO'] = caractericas_vivienda_hogar_2019['ubigeo'].str[:2]

caractericas_vivienda_hogar_2019['departamentos'] = caractericas_vivienda_hogar_2019['COD_DEPARTAMENTO'].map(departamentos)



#agua_potable
caractericas_vivienda_hogar_2019['agua_potable'] = caractericas_vivienda_hogar_2019['agua_potable'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['agua_potable'] = caractericas_vivienda_hogar_2019['agua_potable'].astype(int)
caractericas_vivienda_hogar_2019['agua_potable'] = caractericas_vivienda_hogar_2019['agua_potable'].replace({999: np.nan})

#electricidad

caractericas_vivienda_hogar_2019['electricidad'] = caractericas_vivienda_hogar_2019['electricidad'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['electricidad'] = caractericas_vivienda_hogar_2019['electricidad'].astype(int)
caractericas_vivienda_hogar_2019['electricidad'] = caractericas_vivienda_hogar_2019['electricidad'].replace({999: np.nan})


#telefono
caractericas_vivienda_hogar_2019['telefono'] = caractericas_vivienda_hogar_2019['telefono'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['telefono'] = caractericas_vivienda_hogar_2019['telefono'].astype(int)
caractericas_vivienda_hogar_2019['telefono'] = caractericas_vivienda_hogar_2019['telefono'].replace({999: np.nan})

#celular
caractericas_vivienda_hogar_2019['celular'] = caractericas_vivienda_hogar_2019['celular'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['celular'] = caractericas_vivienda_hogar_2019['celular'].astype(int)
caractericas_vivienda_hogar_2019['celular'] = caractericas_vivienda_hogar_2019['celular'].replace({999: np.nan})

#tv
caractericas_vivienda_hogar_2019['tv'] = caractericas_vivienda_hogar_2019['tv'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['tv'] = caractericas_vivienda_hogar_2019['tv'].astype(int)
caractericas_vivienda_hogar_2019['tv'] = caractericas_vivienda_hogar_2019['tv'].replace({999: np.nan})

#internet
caractericas_vivienda_hogar_2019['internet'] = caractericas_vivienda_hogar_2019['internet'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2019['internet'] = caractericas_vivienda_hogar_2019['internet'].astype(int)
caractericas_vivienda_hogar_2019['internet'] = caractericas_vivienda_hogar_2019['internet'].replace({999: np.nan})

caractericas_vivienda_hogar_2019 = caractericas_vivienda_hogar_2019.dropna(subset=['ubigeo', 'agua_potable', 'electricidad', 'telefono','celular','tv','internet'])

caractericas_vivienda_hogar_2019
```

MODULO SALUD

```python
caractericas_salud_2019 = pd.read_csv('../Data 2019/Salud/Enaho01A-2019-400.csv', encoding='latin1')
caractericas_salud_2019



caractericas_salud_2019 = caractericas_salud_2019[['UBIGEO', 'P4193', 'P4192', 'P4191', 'P401', 'P4195', 'P4194' ]]

caractericas_salud_2019 = caractericas_salud_2019.rename(columns={'ubigeo':'departamento','P4193': 'seguro_eps','P4192': 'seguro_privado',
                             'P4194': 'seguro_policial',
                              'P4191': 'seguro_essalud',
                              'P401': 'enfermedad_cronica','P4195': 'seguro_sis'})

caractericas_salud_2019 = caractericas_salud_2019.rename(columns=str.lower)



caractericas_salud_2019['seguro_eps'] = caractericas_salud_2019['seguro_eps'].replace({" ": "999","2":"0"})
caractericas_salud_2019['seguro_eps'] = caractericas_salud_2019['seguro_eps'].astype(int)
caractericas_salud_2019['seguro_eps'] = caractericas_salud_2019['seguro_eps'].replace({999: np.nan})

#seguro_privado
caractericas_salud_2019['seguro_privado'] = caractericas_salud_2019['seguro_privado'].replace({" ": "999","2":"0"})
caractericas_salud_2019['seguro_privado'] = caractericas_salud_2019['seguro_privado'].astype(int)
caractericas_salud_2019['seguro_privado'] = caractericas_salud_2019['seguro_privado'].replace({999: np.nan})

#seguro_policial
caractericas_salud_2019['seguro_policial'] = caractericas_salud_2019['seguro_policial'].replace({" ": "999","2":"0"})
caractericas_salud_2019['seguro_policial'] = caractericas_salud_2019['seguro_policial'].astype(int)
caractericas_salud_2019['seguro_policial'] = caractericas_salud_2019['seguro_policial'].replace({999: np.nan})

#seguro_essalud
caractericas_salud_2019['seguro_essalud'] = caractericas_salud_2019['seguro_essalud'].replace({" ": "999","2":"0"})
caractericas_salud_2019['seguro_essalud'] = caractericas_salud_2019['seguro_essalud'].astype(int)
caractericas_salud_2019['seguro_essalud'] = caractericas_salud_2019['seguro_essalud'].replace({999: np.nan})

#enfermedad_cronica
caractericas_salud_2019['enfermedad_cronica'] = caractericas_salud_2019['enfermedad_cronica'].replace({" ": "999","2":"0"})
caractericas_salud_2019['enfermedad_cronica'] = caractericas_salud_2019['enfermedad_cronica'].astype(int)
caractericas_salud_2019['enfermedad_cronica'] = caractericas_salud_2019['enfermedad_cronica'].replace({999: np.nan})

#seguro_sis
caractericas_salud_2019['seguro_sis'] = caractericas_salud_2019['seguro_sis'].replace({" ": "999","2.0":"0"})
caractericas_salud_2019['seguro_sis'] = caractericas_salud_2019['seguro_sis'].astype(int)
caractericas_salud_2019['seguro_sis'] = caractericas_salud_2019['seguro_sis'].replace({999: np.nan})

caractericas_salud_2019 = caractericas_salud_2019.dropna(subset=['ubigeo','seguro_sis','enfermedad_cronica','seguro_essalud','seguro_policial','seguro_privado','seguro_eps'])


caractericas_salud_2019
```

MODULO EMPLEO

```python

caractericas_empleo_2019 = pd.read_csv('../Data 2019/Empleo e Ingresos/Enaho01a-2019-500.csv', encoding='latin1')

caractericas_empleo_2019




caractericas_empleo_2019 = caractericas_empleo_2019.rename(columns={'ubigeo':'departamento','D538A1': 'ingresos_mensuales2', 
                              'D524A1': 'ingresos_mensuales1',
                              'OCUPINF': 'formalidad',
                              'P512A': 'tamaño_empresa'
                              })

caractericas_empleo_2019 = caractericas_empleo_2019.rename(columns=str.lower)





#tamaño_empresa

mapeo_tamano_empresa={
    '1': '1',
    '3': '2',
    '4': '3',
    '5': '4',
    '2': '1'
}

caractericas_empleo_2019['tamaño_empresa'] = caractericas_empleo_2019['tamaño_empresa'].str.strip()

caractericas_empleo_2019['tamaño_empresa'] = caractericas_empleo_2019['tamaño_empresa'].replace(mapeo_tamano_empresa)

#formalidad
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].replace({" ": "999","2":"0","NaN":"999"})
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].astype(int)
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].replace({999: np.nan})


#INGRESOS
caractericas_empleo_2019['ingresos_mensuales'] = caractericas_empleo_2019['ingresos_mensuales1'] + caractericas_empleo_2019['ingresos_mensuales2']

caractericas_empleo_2019['ingresos_mensuales'] = pd.to_numeric(caractericas_empleo_2019['ingresos_mensuales'], errors='coerce')

caractericas_empleo_2019['ingresos_mensuales'].value_counts()

ipc_2019 = 1.05

#ingreso mensual deflactado

caractericas_empleo_2019['ingresos_mensuales_deflactados'] = caractericas_empleo_2019['ingresos_mensuales'] / ipc_2019
caractericas_empleo_2019['ingresos_mensuales_deflactados']
caractericas_empleo_2019['ingresos_mensuales_deflactados'] = caractericas_empleo_2019['ingresos_mensuales_deflactados'].replace({" ": "999", "2": "0"})
caractericas_empleo_2019['ingresos_mensuales_deflactados'] = pd.to_numeric(caractericas_empleo_2019['ingresos_mensuales_deflactados'], errors='coerce')
caractericas_empleo_2019['ingresos_mensuales_deflactados'].fillna(np.nan, inplace=True)
caractericas_empleo_2019['ingresos_mensuales_deflactados'] = caractericas_empleo_2019['ingresos_mensuales_deflactados'].round().astype('Int64')


caractericas_empleo_2019 = caractericas_empleo_2019[['ubigeo', 'formalidad', 'tamaño_empresa', 'ingresos_mensuales_deflactados']]

caractericas_empleo_2019 = caractericas_empleo_2019.dropna(subset=['ubigeo', 'formalidad', 'tamaño_empresa', 'ingresos_mensuales_deflactados'])
 
caractericas_empleo_2019
```

### **UNIENDO BASE DE DATOS 2018**

```python
caractericas_miembros_hogar_2019['ubigeo'] = caractericas_miembros_hogar_2019['ubigeo'].astype(str)
caractericas_vivienda_hogar_2019['ubigeo'] = caractericas_vivienda_hogar_2019['ubigeo'].astype(str)

individual_1 = pd.merge(caractericas_miembros_hogar_2019, caractericas_vivienda_hogar_2019, on=['ubigeo'], how='outer')
individual_1 = individual_1.dropna()

```

```python
caractericas_miembros_hogar_2019['ubigeo'] = caractericas_miembros_hogar_2019['ubigeo'].astype(str)
caractericas_educacion_2019['ubigeo'] = caractericas_educacion_2019['ubigeo'].astype(str)

individual_2 = pd.merge(caractericas_miembros_hogar_2019, caractericas_educacion_2019, on=['ubigeo'], how='outer')
individual_2 = individual_2.dropna()
```

```python

caractericas_miembros_hogar_2019['ubigeo'] = caractericas_miembros_hogar_2019['ubigeo'].astype(str)
caractericas_empleo_2019['ubigeo'] = caractericas_empleo_2019['ubigeo'].astype(str)

individual_3 = pd.merge(caractericas_miembros_hogar_2019, caractericas_empleo_2019, on=['ubigeo'], how='outer')
individual_3 = individual_3.dropna()
```

```python
caractericas_miembros_hogar_2019['ubigeo'] = caractericas_salud_2019['ubigeo'].astype(str)
caractericas_salud_2019['ubigeo'] = caractericas_salud_2019['ubigeo'].astype(str)

individual_4 = pd.merge(caractericas_miembros_hogar_2019, caractericas_salud_2019, on=['ubigeo'], how='outer')
individual_4 = individual_4.dropna()
```

```python
dataframes_2019 = [individual_4,individual_3,individual_2,individual_1] 

concatenacion_2019 = pd.concat(dataframes_2019, axis=1)
```
