### **PROCESAMIENTO DEL MÓDULO "CARACTERÍSTICAS DE LA VIVIENDA Y DEL HOGAR"**


```python
# importamos pandas y el documento del 2018 
import pandas as pd
import numpy as np
## Lee el conjunto de datos desde un archivo CSV
caractericas_vivienda_hogar_2018 = pd.read_csv(
    '../data2018/vivienda/634-Modulo01/Enaho01-2018-100.csv',
    encoding='latin1'
)

# una variable de departamento en formato string a partir de la variable UBIGEO. 
# PARA HOGAR2018
caractericas_vivienda_hogar_2018 = caractericas_vivienda_hogar_2018.rename(columns=str.lower)
caractericas_vivienda_hogar_2018['ubigeo'] = caractericas_vivienda_hogar_2018['ubigeo'].astype(str)
caractericas_vivienda_hogar_2018['ubigeo'] = caractericas_vivienda_hogar_2018['ubigeo'].apply(lambda x: x.zfill(6))

def departamentos(ubigeo):
    return ubigeo[:2]

caractericas_vivienda_hogar_2018['departamento'] = caractericas_vivienda_hogar_2018['ubigeo'].apply(departamentos)

caractericas_vivienda_hogar_2018 = caractericas_vivienda_hogar_2018[["año", "conglome", "vivienda", "hogar", "ubigeo", "departamento", "p110a1", "p1121", "p1141", "p1142", "p1143", "p1144"]]


# otorgando nuevos nombres a las variables indicadas
caractericas_vivienda_hogar_2018 = caractericas_vivienda_hogar_2018.rename(columns={'p110a1': 'agua_potable', 
                                      'p1121': 'electricidad',
                                      'p1141': 'telefono',
                                      'p1142': 'celular',
                                      'p1143': 'tv',
                                      'p1144':'internet'})
#departamento
departamentos = {
    "01": "Amazonas",
    "02": "Áncash",
    "03": "Apurímac",
    "04": "Arequipa",
    "05": "Ayacucho",
    "06": "Cajamarca",
    "07": "Callao",
    "08": "Cusco",
    "09": "Huancavelica",
    "10": "Huanuco",
    "11": "Ica",
    "12": "Junin",
    "13": "La Liberta",
    "14": "Lambayeque",
    "15": "Lima",
    "16": "Loreto",
    "17": "Madre de Dios",
    "18": "Moquegua",
    "19": "Pasco",
    "20": "Piura",
    "21": "Puno",
    "22": "San Martin",
    "23": "Tacna",
    "24": "Tumbes",
    "25": "Ucayali"
}
caractericas_vivienda_hogar_2018['ubigeo'] = caractericas_vivienda_hogar_2018['ubigeo'].astype(str)

caractericas_vivienda_hogar_2018['departamentos'] = caractericas_vivienda_hogar_2018['departamento'].map(departamentos)



#agua_potable
caractericas_vivienda_hogar_2018['agua_potable'] = caractericas_vivienda_hogar_2018['agua_potable'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['agua_potable'] = caractericas_vivienda_hogar_2018['agua_potable'].astype(int)
caractericas_vivienda_hogar_2018['agua_potable'] = caractericas_vivienda_hogar_2018['agua_potable'].replace({999: np.nan})

#electricidad

caractericas_vivienda_hogar_2018['electricidad'] = caractericas_vivienda_hogar_2018['electricidad'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['electricidad'] = caractericas_vivienda_hogar_2018['electricidad'].astype(int)
caractericas_vivienda_hogar_2018['electricidad'] = caractericas_vivienda_hogar_2018['electricidad'].replace({999: np.nan})


#telefono
caractericas_vivienda_hogar_2018['telefono'] = caractericas_vivienda_hogar_2018['telefono'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['telefono'] = caractericas_vivienda_hogar_2018['telefono'].astype(int)
caractericas_vivienda_hogar_2018['telefono'] = caractericas_vivienda_hogar_2018['telefono'].replace({999: np.nan})

#celular
caractericas_vivienda_hogar_2018['celular'] = caractericas_vivienda_hogar_2018['celular'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['celular'] = caractericas_vivienda_hogar_2018['celular'].astype(int)
caractericas_vivienda_hogar_2018['celular'] = caractericas_vivienda_hogar_2018['celular'].replace({999: np.nan})

#tv
caractericas_vivienda_hogar_2018['tv'] = caractericas_vivienda_hogar_2018['tv'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['tv'] = caractericas_vivienda_hogar_2018['tv'].astype(int)
caractericas_vivienda_hogar_2018['tv'] = caractericas_vivienda_hogar_2018['tv'].replace({999: np.nan})

#internet
caractericas_vivienda_hogar_2018['internet'] = caractericas_vivienda_hogar_2018['internet'].replace({" ": "999","2":"0"})
caractericas_vivienda_hogar_2018['internet'] = caractericas_vivienda_hogar_2018['internet'].astype(int)
caractericas_vivienda_hogar_2018['internet'] = caractericas_vivienda_hogar_2018['internet'].replace({999: np.nan})

caractericas_vivienda_hogar_2018 = caractericas_vivienda_hogar_2018.dropna(subset=['ubigeo', 'agua_potable', 'electricidad', 'telefono','celular','tv','internet'])


caractericas_vivienda_hogar_2018
```

```python
caractericas_educacion_2018 = pd.read_csv('../data2018/Educación/634-Modulo03/Enaho01a-2018-300.csv'
    ,
    encoding='latin1', )

caractericas_educacion_2018 = caractericas_educacion_2018.rename(columns=str.lower)

caractericas_educacion_2018 = caractericas_educacion_2018.rename(columns={'p301a': 'educacion'})

caractericas_educacion_2018['educacion'] = caractericas_educacion_2018['educacion'].replace({" ": "999", "99": "999"})
caractericas_educacion_2018['educacion'] = caractericas_educacion_2018['educacion'].astype(int)
caractericas_educacion_2018['educacion'] = caractericas_educacion_2018['educacion'].replace({999: np.nan})

caractericas_educacion_2018.loc[(caractericas_educacion_2018['educacion']==1), 'educacion'] = 1
caractericas_educacion_2018.loc[(caractericas_educacion_2018['educacion']==2) | (caractericas_educacion_2018['educacion']==3) | (caractericas_educacion_2018['educacion']==4) | (caractericas_educacion_2018['educacion']==5) | (caractericas_educacion_2018['educacion']==6), 'educacion'] = 2
caractericas_educacion_2018.loc[(caractericas_educacion_2018['educacion']==7) | (caractericas_educacion_2018['educacion']==8) | (caractericas_educacion_2018['educacion']==9) | (caractericas_educacion_2018['educacion']==10), 'educacion'] = 3
caractericas_educacion_2018.loc[(caractericas_educacion_2018['educacion']==11), 'educacion'] = 4


caractericas_educacion_2018 = caractericas_educacion_2018[['año', 'conglome', 'vivienda', 'hogar', 'codperso', 'ubigeo', 'educacion']]
caractericas_educacion_2018
```

### **PROCESAMIENTO DEL MÓDULO ‘EMPLEO’**



```python
import pandas as pd
import numpy as np

# Read the CSV file
caractericas_empleo_2018 = pd.read_csv('../data2018/Empleo e Ingresos/Enaho01a-2018-500.csv', encoding='latin1')

# Select relevant columns
caractericas_empleo_2018 = caractericas_empleo_2018[['AÑO', 'CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO', 'UBIGEO', 'P512A', 'OCUPINF', 'D538A1', 'D524A1']]

# Rename columns
caractericas_empleo_2018 = caractericas_empleo_2018.rename(columns={'AÑO': 'año',
                                                                  'CONGLOME': 'conglome',
                                                                  'VIVIENDA': 'vivienda',
                                                                  'HOGAR': 'hogar',
                                                                  'CODPERSO': 'codperso',
                                                                  'UBIGEO': 'ubigeo',
                                                                  'P512A': 'tamaño_empresa',
                                                                  'OCUPINF': 'formalidad',
                                                                  'D538A1': 'ingresos_totales2',
                                                                  'D524A1': 'ingresos_totales1'})

# Convert 'tamaño_empresa' to string and replace values
caractericas_empleo_2018['tamaño_empresa'] = caractericas_empleo_2018['tamaño_empresa'].astype(str)
caractericas_empleo_2018['tamaño_empresa'] = caractericas_empleo_2018['tamaño_empresa'].str.strip()
mapeo_tamano_empresa = {'1': '1', '3': '2', '4': '3', '5': '4', '2': '1'}
caractericas_empleo_2018['tamaño_empresa'] = caractericas_empleo_2018['tamaño_empresa'].replace(mapeo_tamano_empresa)

# Convert 'formalidad' to integer and replace values
caractericas_empleo_2018['formalidad'] = caractericas_empleo_2018['formalidad'].replace({' ': '999', '2': '0', 'NaN': '999'})
caractericas_empleo_2018['formalidad'] = caractericas_empleo_2018['formalidad'].astype(int)
caractericas_empleo_2018['formalidad'] = caractericas_empleo_2018['formalidad'].replace({999: np.nan})

# Combine 'ingresos_mensuales1' and 'ingresos_mensuales2' to calculate 'ingresos_mensuales'
caractericas_empleo_2018['ingresos_totales'] = pd.to_numeric(caractericas_empleo_2018['ingresos_totales1'], errors='coerce') + pd.to_numeric(caractericas_empleo_2018['ingresos_totales2'], errors='coerce')
caractericas_empleo_2018['ingresos_mensuales'] = caractericas_empleo_2018['ingresos_totales']/12

# Calculate 'ingresos_mensuales_deflactados' and round to integers
ipc_2019 = 1.05
caractericas_empleo_2018['ingresos_mensuales_deflactados'] = caractericas_empleo_2018['ingresos_mensuales'] / ipc_2019
caractericas_empleo_2018['ingresos_mensuales_deflactados'] = caractericas_empleo_2018['ingresos_mensuales_deflactados'].replace({" ": "999", "2": "0"})
caractericas_empleo_2018['ingresos_mensuales_deflactados'] = pd.to_numeric(caractericas_empleo_2018['ingresos_mensuales_deflactados'], errors='coerce')
caractericas_empleo_2018['ingresos_mensuales_deflactados'].fillna(np.nan, inplace=True)
caractericas_empleo_2018['ingresos_mensuales_deflactados'] = caractericas_empleo_2018['ingresos_mensuales_deflactados'].round().astype('Int64')

# Select final columns and drop rows with missing values
caractericas_empleo_2018 = caractericas_empleo_2018[['año', 'conglome', 'vivienda', 'hogar', 'codperso', 'ubigeo', 'tamaño_empresa', 'formalidad', 'ingresos_mensuales', 'ingresos_mensuales_deflactados']]
caractericas_empleo_2018 = caractericas_empleo_2018.dropna(subset=['año', 'conglome', 'vivienda', 'hogar', 'codperso', 'ubigeo', 'tamaño_empresa', 'formalidad', 'ingresos_mensuales', 'ingresos_mensuales_deflactados'])

# Print result
caractericas_empleo_2018
```

```python
caractericas_empleo_2019 = pd.read_csv('../Data 2019/Empleo e Ingresos/Enaho01a-2019-500.csv', encoding='latin1')
# Select relevant columns
caractericas_empleo_2019 = caractericas_empleo_2019[['AÑO', 'CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO', 'UBIGEO', 'P512A', 'OCUPINF', 'D538A1', 'D524A1']]

# Rename columns
caractericas_empleo_2019 = caractericas_empleo_2019.rename(columns={'AÑO': 'año',
                                                                  'CONGLOME': 'conglome',
                                                                  'VIVIENDA': 'vivienda',
                                                                  'HOGAR': 'hogar',
                                                                  'CODPERSO': 'codperso',
                                                                  'UBIGEO': 'ubigeo',
                                                                  'P512A': 'tamaño_empresa',
                                                                  'OCUPINF': 'formalidad',
                                                                  'D538A1': 'ingresos_totales2',
                                                                  'D524A1': 'ingresos_totales1'})

# Convert 'tamaño_empresa' to string and replace values
caractericas_empleo_2019['tamaño_empresa'] = caractericas_empleo_2019['tamaño_empresa'].astype(str)
caractericas_empleo_2019['tamaño_empresa'] = caractericas_empleo_2019['tamaño_empresa'].str.strip()
mapeo_tamano_empresa = {'1': '1', '3': '2', '4': '3', '5': '4', '2': '1'}
caractericas_empleo_2019['tamaño_empresa'] = caractericas_empleo_2019['tamaño_empresa'].replace(mapeo_tamano_empresa)

# Convert 'formalidad' to integer and replace values
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].replace({' ': '999', '2': '0', 'NaN': '999'})
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].astype(int)
caractericas_empleo_2019['formalidad'] = caractericas_empleo_2019['formalidad'].replace({999: np.nan})

# Combine 'ingresos_mensuales1' and 'ingresos_mensuales2' to calculate 'ingresos_mensuales'
caractericas_empleo_2019['ingresos_totales'] = pd.to_numeric(caractericas_empleo_2019['ingresos_totales1'], errors='coerce') + pd.to_numeric(caractericas_empleo_2019['ingresos_totales2'], errors='coerce')
caractericas_empleo_2019['ingresos_mensuales'] = caractericas_empleo_2019['ingresos_totales']/12
# No es necesario deflactar, puesto que se está en el mismo año 2019

# Select final columns and drop rows with missing values
caractericas_empleo_2019 = caractericas_empleo_2019[['año', 'conglome', 'vivienda', 'hogar', 'codperso', 'ubigeo', 'tamaño_empresa', 'formalidad', 'ingresos_mensuales']]
caractericas_empleo_2019 = caractericas_empleo_2019.dropna(subset=['año', 'conglome', 'vivienda', 'hogar', 'codperso', 'ubigeo', 'tamaño_empresa', 'formalidad', 'ingresos_mensuales'])

# Print result
caractericas_empleo_2019
```
